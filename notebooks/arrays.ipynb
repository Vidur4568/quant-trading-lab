### How long to double our investment

```python
capital = 1000
t = 1

2 * capital == capital * 1.05 ** t
```

```python
2 == (1.05) ** t
```

```python
np.log(2) == t * np.log(1.05) 
```

```python
t = np.log(2) / np.log(1.05)
```

```python
t 
```

```python
capital * 1.05 ** t
```
